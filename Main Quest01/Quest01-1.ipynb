```python
import random as r                 #random 모듈을 사용하기 위해 임포트하고 편의성을 위해 r로 as를 줌

class Account:                     #Account 클래스 생성
    bank_name = 'SC은행'           #은행명은 SC은행으로 통일
    num_accounts = 0               # 계좌의 수는 0으로 초기화

    def __init__(self, owner, balance):   #생성자는 예금주 owner와 초기 잔액인 balance를 받음
        self.bank_name = Account.bank_name  #클래스의 은행으로 지정
        self.owner = owner                #생성자의 예금주는 예금주로 지정
        self.balance = balance            #위쪽 설명과 동일
        self.account_number = self.generate_account_number()  #계좌 번호는 해당 인스턴스 메서드에서 가져옴
        self.deposit_count = 0            # 이자를 위한 입금 횟수는 0으로 초기화 설정
        self.deposit_records = []         # 입금 내역을 위해 빈 리스트 생성
        self.withdraw_records = []        # 출금 내역을 위해 빈 리스트 생성
        Account.num_accounts += 1            # 계좌의 수를 생성마다 하나씩 늘려줌

    def generate_account_number(self):           #랜덤 계좌 번호 생성 인스턴스 메서드
        number = str(r.randint(0, 999)).zfill(3)     #랜덤으로 0~999의 숫자를 지정하고 혹시나 1자리 숫자나 2자리 숫자가 나올 수 있기 때문에 zfill함수를 이용 무조건 3자리가 나오게 함 ex) 002
        branch = str(r.randint(0, 99)).zfill(2)      # zfill 함수 때문에 무조건 2자리 숫자가 나옴
        account = str(r.randint(0, 999999)).zfill(6) # zfill 함수 때문에 무조건 6자리 숫자가 나옴
        return f'{number}-{branch}-{account}'      #변수3개 사이에 '-'를 줌으로써 3자리-2자리-6자리인 계좌가 생성되게 함

    def deposit(self, amount):         #입금 인스턴스 메서드
        if amount < 1:                 # 금액이 1원보다 작다면
            print('최소 입금 금액은 1원 이상이어야 합니다.')
            print()
        else:                         # 아니라면
            self.balance += amount    #잔고에 지정한 금액 값을 더함
            self.deposit_count += 1   # 입금 횟수를 카운트함, 이자 카운트를 위한 코드
            self.deposit_records.append(amount)  #입금 기록에 추가함, 입금 내역서를 위한 코드
            print(f'{self.owner}님의 계좌에 {Account.format_balance(amount)}원이 입금되었습니다.\n 현재 잔액 : {Account.format_balance(self.balance)}원')
            print()
            if self.deposit_count == 5:    # 만약 입금 횟수가 5번이라면
                self.pay_interest()        # 이자 발생 인스턴스 메서드 활성화
                self.deposit_count = 0     # 다시 입금 횟수를 0번으로 초기화 이래야 5번 마다 이자가 계속 발생함

    def withdraw(self, amount):        # 출금 인스턴스 메서드
        if amount > self.balance:      # 잔고에 있는 금액보다 출금금액이 많다면
            print('출금 금액이 잔고보다 많습니다.')
            print()
        else:                          # 아니라면
            self.balance -= amount     # 잔고에서 지정한 금액을 뺌
            self.withdraw_records.append(amount)    # 출금 기록에 추가함, 출금 내역서를 위한 코드
            print(f'{self.owner}님의 계좌에서 {Account.format_balance(amount)}원이 출금되었습니다.\n 현재 잔액 : {Account.format_balance(self.balance)}원')
            print()

    def display_info(self):          # 정보 출력 인스턴스 메서드
        print('은행명:', self.bank_name)
        print('예금주:', self.owner)
        print('계좌번호:', self.account_number)
        print('잔고:', Account.format_balance(self.balance), '원')     # 포맷팅을 사용해서 천단위로 ','가 나타나게 함 그것을 사용하는 정적 메서드 self로 쓰지 않은 이유는
        print()                                                        # Account 클래스에는 속하지만 특정 인스턴스와는 관련없는 독립적인 메서드 이므로 Account클래스를 바로 호출

    def format_balance(balance):      # 정적 메서드, balance의 정보를 받아서 처리함
        return '{:,}'.format(balance)  #{:,} 포멧팅으로 천단위로 나눠서 ,을 써줌

    def pay_interest(self):             # 이자 발생 인스턴스 메서드
        interest = int(self.balance * 0.01)      # 이자는 1%발생을 위해 잔고에 0.01을 곱하고 float가 아닌 딱 떨어지게 하기위해 int를 사용
        self.balance += interest          # 잔고에 이자를 추가함
        print(f'{self.owner}님의 계좌에 이자 {interest}원이 입금되었습니다.\n 현재 잔액 : {Account.format_balance(self.balance)}원')
        print()

    @classmethod
    def get_account_num(cls):    # 계좌의 수 카운트 클래스 메서드 클래스 메서드에서는 인스턴스 메서드에서 변수를 접근하는 것이 아닌
                                 #클래스 자체에서 변수를 가져오기에 Account의 변수를 조작함
        return cls.num_accounts


    def deposit_history(self):        # 입금 내역 인스턴스 메서드
        print('입금 내역:')
        for amount in self.deposit_records:   # 입금 기록을 반복해 출력하게함
            print(Account.format_balance(amount), '원')
        print()         # 줄 바꾸기용

    def withdraw_history(self):        # 출금 내역 인스턴스 메서드
        print('출금 내역:')
        for amount in self.withdraw_records:   # 출금 기록을 반복해 출력하게함
            print(Account.format_balance(amount), '원')
        print()

    def VIP(accounts):           #잔고에 백만원이 넘는 계좌를 가진 사람만 보여주는 정적 메서드
        for account in accounts:   #accounts안에 있는 리스트를 반복해서
            if account.balance >= 100000:   # 잔고가 백만원 이상이면
                account.display_info()      # 그 사람의 정보를 출력
                print()


accounts = []    # 객체 생성 후 저장을 위한 빈 리스트
accounts.append(Account('yohan', 0))       # 계좌 추가
accounts.append(Account('dasom', 5000))
accounts.append(Account('Red', 1000000))
accounts.append(Account('Black', 55450010))

print(f'SC은행에 있는 계좌의 갯수는 {Account.get_account_num()}개 입니다.')
print('-'*30)

print('VIP 계좌 목록\n')
Account.VIP(accounts)      # VIP 정보 출력
print('-'*30)

account1 = accounts[1]         # 저장한 리스트에서 2번째 계좌를 변수에 저장

account1.display_info()
account1.deposit(1000)
account1.deposit(0)
account1.deposit(5000)
account1.deposit(4000)
account1.deposit(250000)
account1.deposit(5000)
account1.deposit(500000)
account1.deposit(5000)
account1.deposit(10000)
account1.deposit(5000)
account1.deposit(25000)
account1.deposit(100000)
account1.deposit(500000)
account1.withdraw(3000)
account1.withdraw(72000000)
account1.withdraw(250000)
account1.deposit_history()
account1.withdraw_history()

print('VIP 계좌 목록\n')
Account.VIP(accounts)
print('-'*30)
```
